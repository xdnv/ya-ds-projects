#### Адаптация части колонок к нотации Snake Case
Это не мешает анализу, но повышает наглядность.

```python
# переименование столбцов датасета
column_rename = {
    'cityCenters_nearest': 'city_сenters_nearest',
    'parks_around3000': 'parks_around_3000',
    'ponds_around3000': 'ponds_around_3000',
    'studio': 'is_studio',
    'open_plan': 'is_open_plan'
}
df = df.rename(columns=column_rename)
```

#### Преобразование типов колонок

```python
# Преобразуем поле first_day_exposition из строки в дату
#Данное поле потребуется в дальнейших операциях с датами.
df.first_day_exposition = pd.to_datetime(df.first_day_exposition, format='%Y-%m-%dT%H:%M:%S')
```

```python
#стоимость квартиры, тем более в объявлении, в копейках обычно не измеряется. В анализе будет более наглядно.
df['last_price'] = df['last_price'].astype('int64')
#print(df.sort_values('last_price',ascending='False')['last_price'].unique(),'\n')

# дискретные вещественные значения
df['balcony'] = df['balcony'].astype('Int64')
df['floors_total'] = df['floors_total'].astype('Int64')
df['ponds_around_3000'] = df['ponds_around_3000'].astype('Int64')
df['parks_around_3000'] = df['parks_around_3000'].astype('Int64')

# в колонке days_exposition есть дробные значения, поэтому дополнительно округляем
df['days_exposition'] = round(df['days_exposition'],0).astype('Int64')

# метры
df['airports_nearest'] = df['airports_nearest'].astype('Int64')
df['city_сenters_nearest'] = df['city_сenters_nearest'].astype('Int64')
df['parks_nearest'] = df['parks_nearest'].astype('Int64')
df['ponds_nearest'] = df['ponds_nearest'].astype('Int64')

df.info()
display(df.head(5))
```

#### Обработка колонок, критичных для последующих расчётов

```python
# Убираем пустые значения поля balcony
print('balcony', df.sort_values('balcony')['balcony'].unique())
df['balcony'] = df['balcony'].fillna(0)
print('balcony', df.sort_values('balcony')['balcony'].unique())
```

```python
# Убираем пустые значения поля is_apartment
print('is_apartment', df.sort_values('is_apartment')['is_apartment'].unique(),'\n')
print(df.groupby('is_apartment')['is_apartment'].count(),'\n')


# Пятьдесят значений на 23000 строк - это, в общем и целом, повод данные удалить.
# Вероятно, данная аналитика появилась недавно и большая часть объектов вносилась без её указания.
# Пока что преобразуем в булево без пустых значений, т.к. данное поле может стать актуальным при аназизе выбросов по площади:
#    например, апартаменты с больешй вероятностью могут иметь площадь и 7м2, и 400 м2.

df['is_apartment'] = df['is_apartment'].fillna(False)
print('is_apartment', df.sort_values('is_apartment')['is_apartment'].unique(),'\n')
print(df.is_apartment.value_counts())
```

```python
# Для остальных числовых колонок заменим пустые значения на "магическое" -1
# по мере продвижения анализа такие данные можно будет удалять или заполнять иными, более подходящими данными
for column in ['floors_total',
               'living_area','kitchen_area',
               'airports_nearest','city_сenters_nearest',
               'parks_around_3000','parks_nearest',
               'ponds_around_3000','ponds_nearest']:
    df[column] = df[column].fillna(-1)
```

```python
# Исключение дублей locality_name, связанных с регистром
# создаём колонку в нижнем регистре для исключения дублей по регистру и сразу же избавляемся от Ё
df['locality_name_lc'] = df.locality_name.str.lower().str.replace('ё', 'е', regex=False)

# удаляем типы населённых пунктов
df['locality_name_lc'] = df.locality_name_lc.str.replace('поселок городского типа', '', regex=False)
df['locality_name_lc'] = df.locality_name_lc.str.replace('город', '', regex=False)
df['locality_name_lc'] = df.locality_name_lc.str.replace('деревня', '', regex=False)
df['locality_name_lc'] = df.locality_name_lc.str.replace('поселок', '', regex=False)

# избавляемся от множественных пробелов (если есть)
df['locality_name_lc'] = df.locality_name_lc.str.replace('  ', ' ', regex=False).str.replace('  ', ' ', regex=False)

# удаляем лидирующие и оконечные whitespaces
df['locality_name_lc'] = df.locality_name_lc.str.strip()

print(df.locality_name_lc.value_counts())
```

```python
# Округление стоимости квартир
# данный столбец не имеет пропусков
# Изучим столбец last_price и округлим его значения до сотен тысяч
df['last_price'] = df.last_price.round(decimals=-5)
# Создадим столбец last_price_mio в миллионах рублей - для исследовательского анализа
df['last_price_mio'] = round(df.last_price/1000000,1).astype('int')
print(df.last_price.value_counts(),'\n')
print(df.last_price_mio.value_counts())
```

```python
#Удалим выбросы и искажения данных модели

#вычислим медианное значение и заполним им значения с NaN
#таких записей более 10% выборки, их удаление нежелательно
##############################df['days_exposition'] = df['days_exposition'].fillna(df.days_exposition.median())

# Квартиры, продаваемые в течение сверхдлинного срока (примем - более года) вызывают вопросы
# скорее всего это неликвидные варианты или фейки, призванные повлиять на стоимость других продаваемых в том же месте (доме, районе) квартир.
# удалим эти данные из датасета
display(df.query('days_exposition > 365'))
df = df.query('days_exposition <= 365').reset_index(drop=True)

df.info()
```

```python
# выявлено 143 квартиры дешевле 1 миллиона рублей. Это специфические квартиры в регионах и они могут негативно влиять на модель.
# Удалим эти строки
display(df.query('last_price < 1000000 '))

# выявлено 18 квартир дороже 100 миллионов рублей. Также удалим эти строки
display(df.query('last_price > 100000000 '))
df = df.query('last_price >= 1000000 and last_price <= 100000000').reset_index(drop=True)
```

```python
# выявленs редкие квартиры с существенно большей, чем у остальных, площадью - 25 строк для среза 350+.
# Классифицируем её как выброс и удалим.
display(df.query('total_area > 350'))
df = df.query('total_area <= 350').reset_index(drop=True)
```

```python
# выявлены квартиры с нулём комнат - 182 строки.
# Проверка показывает, что по площади они явно - однокомнатные.
# Если бы записей было существенно больше, имело бы смысл заполнить цифрой 1. В текущем задании просто удалим их.
df_temp = df.query('rooms == 0')
display(df_temp)
print(df_temp.total_area.value_counts())
df = df.query('rooms > 0 and rooms < 11').reset_index(drop=True)
```

```python
#вычислим медианное значение и заполним им значения с NaN
#таких записей более 30% выборки, их удаление нежелательно
df['ceiling_height'] = df['ceiling_height'].fillna(df.ceiling_height.median())

# Высота потолков выше 10 метров вызывает большие вопросы: а квартиры ли это вообще?
# всего 22 записи, удаляем их
display(df.query('ceiling_height > 10'))

# Три записи с потолком ниже 2м - удаляем
display(df.query('ceiling_height < 2'))
df = df.query('ceiling_height <= 10 and ceiling_height >= 2').reset_index(drop=True)
```

```python
# 39 записей без указания населённого пункта. Также удалим.
display(df.loc[df['locality_name'].notna()])

df = df.loc[df['locality_name'].notna()].reset_index(drop=True)
```

## Расчёты и добавление результатов в таблицу

### Шаг 3. Расчёт основных параметров и добавление их в таблицу
* цену квадратного метра;
* день недели, месяц и год публикации объявления;
* этаж квартиры; варианты — первый, последний, другой;
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
# вычисляем цену квадратного метра в целых рублях
def set_price_sqm(row):
    return int(round(row['last_price']/row['total_area'],0))

df['price_sqm'] = df.apply(set_price_sqm, axis=1)

# вычисляем день недели, месяц и год публикации объявления
df['exposition_year'] = pd.DatetimeIndex(df['first_day_exposition']).year
df['exposition_month'] = pd.DatetimeIndex(df['first_day_exposition']).month
df['exposition_weekday'] = df['first_day_exposition'].dt.weekday # здесь именно день недели, не число месяца


# вычисляем этаж квартиры; варианты — первый, последний, другой
def set_floor_type(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if (floor == 1):
        return 'первый'
    elif (floor == floors_total):
        return 'последний'
    else:
        return 'другой'

df['floor_type'] = df.apply(set_floor_type, axis=1)


# вычисляем соотношение жилой и общей площади
df['living_room_ratio'] = (df['living_area'] / df['total_area']).round(2)

# вычисляем отношение площади кухни к общей
df['kitchen_room_ratio'] = (df['kitchen_area'] / df['total_area']).round(2)


display(df.head(5))
```

"Удалённость от центра" имеет слабую обратную корреляцию. Вероятно, это обусловлено редкостью и не самым хорошим состоянием исторических квартир в самом центре.

```python
# Тут я создам две случайные выборки из 100 чисел и посчитаю коэффициент корреляции
import numpy as np
np.random.seed(38)
a = np.random.random(100)
b = np.random.random(100)
np.corrcoef(a, b)
```

```python
# Пойду дальше - сгенерирую этот эксперимент 10000 раз и построю гистограмму коэффициентов корреляции
N = 10000
result = []
for i in range(N):
    a = np.random.random(100)
    b = np.random.random(100)
    coef = np.corrcoef(a, b)[0, 1]
    result.append(coef)
plt.hist(result);
```
